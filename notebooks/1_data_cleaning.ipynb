```python
# Importing Libraries
import pandas as pd
import numpy as np
import os
import warnings
warnings.filterwarnings("ignore")
```

```python
# Loading Dataset
#Defining File Path
data_path = "../data/shop.csv"

# Loading dataset
df = pd.read_csv(data_path)

# Showing first lows
df.head()
```

```python
# Previewing all column names
df.columns
```

```python
# Standarding Column names
df.columns = [col.strip().replace(" ", "_").lower() for col in df.columns]
```

```python
# Previweing again
df.columns
```

```python
# Droping columns we wont use in model
cols_to_drop = [
    'campaign_name',
    'ad_set_name',
    'delivery_status',
    'delivery_level',
    'attribution_setting',
    'quality_ranking',
    'engagement_rate_ranking',
    'conversion_rate_ranking',
    'reporting_starts',
    'reporting_ends'
]

df.drop(columns=cols_to_drop, inplace= True)
df.head()
```

```python
# Replace 0 impressions to avoid division by zero
df['impressions'] = df['impressions'].replace(0, np.nan)

# Create CTR column
df['ctr'] = (df['results'] / df['impressions']) * 100

# Drop rows with NaN CTR
df.dropna(subset=['ctr'], inplace=True)

# Reset index
df.reset_index(drop=True, inplace=True)

df[['results', 'impressions', 'ctr']].head()
```

```python
# Save to processed folder
processed_path = "../data"
os.makedirs(processed_path, exist_ok=True)
df.to_csv(os.path.join(processed_path, "shop_clean.csv"), index=False)
```
